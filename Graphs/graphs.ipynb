```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

# Example data
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
z = [5, 10, 15, 20, 25]

# Create a 3D plot
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.scatter3D(x, y, z, c=z, cmap='Blues')

# Show the plot
plt.show()
```
